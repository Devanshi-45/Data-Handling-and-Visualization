```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import textwrap
data = pd.read_csv("jobs_in_data.csv")
data.head()
data.shape
data.columns
data.dtypes
data.isnull().sum()
sns.set(style='white')
experience_counts = data['experience_level'].value_counts()
company_size_counts = data['company_size'].value_counts()
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16, 12),facecolor='beige',gridspec_kw={ 'hspace': 0.3})
fig.patch.set_facecolor('beige') 
axes[0, 0].hist(data['salary_in_usd'], bins=15, color='yellow', edgecolor='black')
axes[0, 0].set_title('Salary Distribution for different data jobs',fontweight='bold')
axes[0, 0].set_xlabel('Salary')
axes[0, 0].set_ylabel('Frequency')
axes[0,0].set_facecolor('beige')
sns.countplot(x='experience_level', data=data, palette='magma',ax=axes[0, 1])
axes[0,1].set_title('Distribution of Experience Levels',fontweight='bold')
axes[0,1].set_xlabel('Experience Levels')
axes[0,1].set_ylabel('Count')
axes[0,1].set_facecolor('beige')
least_frequent = company_size_counts.idxmin()
explode = [0.4 if size == least_frequent else 0 for size in company_size_counts.index]
wedges, texts, autotexts = axes[1,0].pie(company_size_counts, autopct='%1.2f%%', startangle=45,explode=explode,colors=['pink','violet','purple'],shadow=True)
axes[1,0].legend(wedges, company_size_counts.index, title='Company Size', loc='lower center', bbox_to_anchor=(-0.04, -0.04))
axes[1,0].set_title('Distribution of Company Sizes',fontweight='bold')
work_setting_counts = data['work_setting'].value_counts().reset_index()
sns.barplot(x='work_setting',y='index', data=work_setting_counts, palette='viridis',width=0.4,ax=axes[1,1])
axes[1,1].set_xlabel('Number of Employees')
axes[1,1].set_ylabel('Work Setting')
axes[1,1].set_title('Number of Employees in Different Work Settings',fontweight='bold')
axes[1,1].set_facecolor('beige')
author_box= dict(boxstyle='round',facecolor='pink',alpha=0.4,edgecolor='purple')
plt.text(0.50,-0.03,'Author : Devanshi Wadkar',fontsize=18,color='Black', ha='center', va='center', transform=fig.transFigure,bbox=author_box)
plt.text(0.50,-0.06,'Student ID : 22065576',fontsize=18,color='Black', ha='center', va='center', transform=fig.transFigure,bbox=author_box)
plt.suptitle('Jobs in Data', fontsize=26,fontweight='heavy',backgroundcolor='Purple',color='White')
r_text =''' The highest salary for any data related jobs lies between 100K-180K with the most being 150000. 
From the countplot we can clearly see there are much positions given to the seniors of upto 6800 ,
also the companies that have provided jobs in the field of data are given by Medium-sized Companies and least being small-sized companies.
The data jobs have the most number of employees working in office and least being in hybrid mode.'''
text = textwrap.fill(r_text,width=125,initial_indent='',subsequent_indent='')
plt.text(0.5,0.035,text,fontsize=16,color='Black', ha='center', va='center', transform=fig.transFigure)
plt.show()

```
